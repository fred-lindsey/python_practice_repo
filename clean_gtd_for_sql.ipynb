```python
import pandas as pd
```

```python
df = pd.read_csv('globalterrorismdb_0522dist.csv')
df.head()
```

```python
df.isna().sum().sort_values(ascending=False).head(15)
```

```python
def clean_gt_df(df, level):
    cols_to_drop = []
    for col in list(df.columns):
        if df[col].isna().sum() > level:
            cols_to_drop.append(col)
    df = df.drop(columns = cols_to_drop)
    print(cols_to_drop)
    return df
```

```python
fiveKNulls = clean_gt_df(df, 5000)
fiveKNulls
```

```python
fiveKNulls.isna().sum().sort_values(ascending=False)
```

```python
len(fiveKNulls)
```

```python
# drop null rows at this point
fiveKNulls = fiveKNulls.dropna()
len(fiveKNulls)
```

```python
fiveKNulls.isna().sum().head(2)
```

```python
fiveKNulls.to_csv('gtd_no_nulls.csv')
```
